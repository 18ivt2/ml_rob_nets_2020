```python
def get_last_name(name):
    return name.split(',')[0].strip()

last_names = pass_data['Name'].apply(get_last_name)
last_names.head()
```

```python
pass_data['Last_name'] = last_names
pass_data.head()
```

```python
pass_data.drop('Last_name', axis = 1, inplace = True)
pass_data.head()
```
