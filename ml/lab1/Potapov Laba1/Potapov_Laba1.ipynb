8) Определите коэффициент корреляции Пирсона и Спирмена между стоимостью автомобиля и типом кузова. Оцените значение p-value. Постройте плотность распределения признаков.

```python
from scipy.stats import pearsonr
from scipy.stats import spearmanr
from sklearn.preprocessing import LabelEncoder
```

```python
vehicle_type = pd.Series(LabelEncoder().fit_transform(fil_cars.vehicleType))
pearson = pearsonr(fil_cars.price,vehicle_type)
spearman = spearmanr(fil_cars.price,vehicle_type)
```

```python
print('Коэффициент Пирсона: ', pearson)
print('Коэффициент Спримена: ', spearman)
if spearman[1] < 0.05:
    print('Приемлимое значение p-value')
else:
    print('Не приемлимое значение p-value')
fil_cars.price.plot.kde()
```
