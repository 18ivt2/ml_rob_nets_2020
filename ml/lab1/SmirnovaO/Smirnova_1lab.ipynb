#### 6. Какое самое популярное женское имя на корабле? Извлеките из полного имени пассажира (колонка Name) его личное имя (First Name). Попробуйте вручную разобрать несколько значений столбца Name и выработать правило для извлечения имен, а также разделения их на женские и мужские.

```python
import re
def get_first_name(name):
    
    stroka = re.search('^[^,]+, (.*)', name)
    if stroka:
        name = stroka.group(1)
    stroka = re.search('\(([^)]+)\)', name)
    if stroka:
        name = stroka.group(1)

    name = re.sub('(Miss\. |Mrs\. |Ms\. )', '', name)
    name = name.split(' ')[0].replace('"', '')
    return name


last_names=titanic_df[titanic_df['Sex']=='female']['Name'].map(get_first_name)
last_names_count= last_names.value_counts()
last_names_count.head()
```

#### 10. Постройте столбчатую диаграмму количества людей: мужчины, женщины, дети.

```python
child_df = titanic_df[titanic_df.Age<18]
male_df = titanic_df[(titanic_df.Sex == 'male')&(titanic_df.Age>=18)]
female_df = titanic_df[(titanic_df.Sex == 'female')&(titanic_df.Age>=18)]
child_df['TypePeople'] = 'Child'
male_df['TypePeople'] = 'Male'
female_df['TypePeople'] = 'Female'
combined_df = pd.concat([child_df, male_df, female_df], ignore_index=True)
combined_df.groupby(['TypePeople'])['PassengerId'].count().plot(kind = 'bar')
combined_df['PassengerId'].count()


```
