```python
import pandas as pd
autos_data = pd.read_csv('autos.csv', encoding='iso-8859-1')
autos_data
```



1. Удалите столбцы, ценность которых для оценки стоимости машины низка.


```python
autos_data_price_influence = autos_data.drop(
    ['seller',
     'abtest', 
     'dateCrawled',
     'offerType',
     'postalCode',
     'dateCreated',
     'nrOfPictures', 
     'lastSeen'],
    axis = 1)
autos_data_price_influence
```

2. Удалите повторяющиеся строки, строки содержащие пропуски в данных.

```python
autos_data_clear_duplicates = autos_data_price_influence.drop_duplicates(
    ['name',
     'price',
     'vehicleType',
     'yearOfRegistration',
     'gearbox',
     'powerPS',
     'model',
     'kilometer',
     'monthOfRegistration',
     'fuelType',
     'notRepairedDamage'
    ])
autos_data_clear_duplicates
```

```python
autos_data_clear_duplicates = autos_data_clear_duplicates[
    autos_data_clear_duplicates.notnull().all(axis=1)
]
autos_data_clear_duplicates
```

3. Удалите данные в строках, выходящие за некоторые пределы (год регистрации, цена, мощность двигателя).

```python

autos_data_limits = autos_data_clear_duplicates[
    (autos_data_clear_duplicates.price >= 1000)&
    (autos_data_clear_duplicates.price <= 8000)&
    (autos_data_clear_duplicates.yearOfRegistration >= 1980 ) & 
    (autos_data_clear_duplicates.yearOfRegistration <= 2000) & 
    (autos_data_clear_duplicates.powerPS >= 200) & 
    (autos_data_clear_duplicates.powerPS <= 500)
]
autos_data_limits
```

7. Постройте карту корреляций между признаками. Выведите также числовые значения признаков.

```python
import seaborn as sn
correlation  = autos_data_length.corr()
heatmap = sn.heatmap(correlation, annot=True)
heatmap
correlation
```

```python
from scipy.stats import pearsonr
from scipy.stats import spearmanr
from sklearn.preprocessing import LabelEncoder

lblnc = LabelEncoder()
vehicleCoder = pd.Series(lblnc.fit_transform(autos_data_length.vehicleType))
pearsoncoef = pearsonr(autos_data_length.price,vehicleCoder)
spearmancoef = spearmanr(autos_data_length.price,vehicleCoder)

print("Коэффициент корреляции Пирсона: ",pearsoncoef)
print("Коэффициент корреляции Спирмена: ",spearmancoef)

if (spearmancoef[1] < 0.05):
    print("Значимая вероятность ошибки")
else:
    print("Не значимая")

a_price = pd.Series(autos_data_length.price)
a_price.plot.kde()
```
