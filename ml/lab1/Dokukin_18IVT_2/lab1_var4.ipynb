### 1) Постройте графики количества осадков по годам в разных штатах

```python
df_graph = df.groupby(['SUBDIVISION'])
fig,ax = plt.subplots(figsize=(8,6))
for label, df_new in df_graph:
    df_new.plot(kind='line', ax=ax, label=label, x='YEAR', y='ANNUAL', figsize=(20,10))
plt.show()
```

### 2) Постройте графики количества осадков по месяцам по всем годам кумулятивно

```python
df_cum = df[['JAN','FEB','MAR','APR','MAY','JUN','JUL','AUG','SEP','OCT','NOV','DEC']]
df_cum = df_cum.cumsum()
df_cum.plot(figsize=(8,8))
plt.show()
```

### 9) Вычислите корреляцию Пирсона и Спирмена между признаками NOV и MAR. Сравните полученные величины корреляции. Постройте гистограммы для указанных признаков. Оцените значение p-value

```python
from scipy.stats import pearsonr as prs
prs(df['NOV'], df['MAR'])
```

```python
from scipy.stats import spearmanr as spr
spr(df['NOV'],df['MAR'], nan_policy = 'omit')
```

Из того, что значение pvalue мало, следует, что мы отвергаем нулевую гипотезу. Поскольку коэффициент корреляции между признаками очень мал, можно сделать вывод, что корреляция слабо выражена.

```python
df['NOV'].plot(kind='hist')
plt.show()
```
