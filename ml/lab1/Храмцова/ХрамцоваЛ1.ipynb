9. Вычислите корреляцию Пирсона и Спирмена между признаками Gender и Missions. Сравните полученные величины корреляции. Постройте гистограммы для указанных признаков. Оцените значение p-value.

```python
from scipy.stats import pearsonr
from scipy.stats import spearmanr
from sklearn.preprocessing import LabelEncoder

```

```python
astronauts_df= astronauts_df.dropna(subset=['Gender', 'Missions'])
G = LabelEncoder().fit_transform(astronauts_df.Gender.astype(str))
M =  LabelEncoder().fit_transform(astronauts_df.Missions.astype(str))
Pearson = pearsonr(pd.Series(G), pd.Series(M))
Spearman = spearmanr(pd.Series(G), pd.Series(M))
print("Pearson: ", Pearson)
print("Spearman: ", Spearman)


```

```python
if (Spearman[1]<0.05):
    print("p-value is good")
else:
    print("p-value is not good")
```
