```python
# Кольчугин Илья 17-АС.
# Лабораторная работа №1.
# Вариант 3.

import pandas as pd
autos_data = pd.read_csv('autos.csv', encoding='iso-8859-1')
```

```python
#1. Удалите столбцы, ценность которых для оценки стоимости машины низка.
autos_data_drop = autos_data.drop(['seller', 'dateCrawled','offerType','postalCode','dateCreated','nrOfPictures'], axis = 1)
autos_data_drop
```

```python
#2. Удалите повторяющиеся строки
au_data_drop = autos_data_drop.drop_duplicates()
```

```python
#...строки содержащие пропуски в данных. За пропуск в данных я принял 'NaN'.
import numpy as np
au_data_drop = au_data_drop[au_data_drop.notnull().all(axis=1)]
```

```python
#...Выведите размер набора данных до и после удаления.
#До удаления дубликатов и повторов:
autos_data_drop
```

```python
#После удаления дубликатов и повторов:
au_data_drop
```

```python
#3. Удалите данные в строках, выходящие за некоторые пределы (год регистрации, цена, мощность двигателя). 
condition_au = au_data_drop[(au_data_drop.price < 10000) & (au_data_drop.price != 0) | (au_data_drop.price > 20000) | (au_data_drop.yearOfRegistration < 2005 ) | (au_data_drop.yearOfRegistration > 2010) | (au_data_drop.powerPS > 200) | (au_data_drop.powerPS < 150)  & (au_data_drop.powerPS != 0)]
au_data_limits = au_data_drop.drop(condition_au.index)
```

```python
# 4. Заполнить пропущенные данные в строковых полях. Заполнил (заменил) все нулевые значения на значением 1337.
replace_data_au = au_data_limits.replace(0, 1337)
```
